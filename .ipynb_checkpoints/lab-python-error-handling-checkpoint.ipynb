def initialize_inventory(products):

   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

4. Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.


````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

def get_customer_orders(inventory):
    number_customer_orders = 0
    valid_number = False
    while not valid_number:
        try:
            number = int(input("Enter the number of customer orders: "))
            if number < 0:
                raise ValueError("Invalid number! Please enter a non-negative value.")
            valid_number = True
        except ValueError as error:
            print(f"Error: {error}")
    number_customer_orders += number
    
    customer_orders = set()
    for i in range(number_customer_orders):
        valid_name = False
        while not valid_name:
            try:
                name = input("Enter the name of a product that a customer wants to order: ")   
                if name not in inventory.keys() or inventory[name]==0:
                    raise ValueError("Invalid product name! Please enter a name of a product that is available in the inventory.")
                valid_name = True
            except ValueError as error:
                print(f"Error: {error}")
        customer_orders.add(name)
    return customer_orders

def update_inventory(customer_orders, inventory):
    return {key: value-1 if key in customer_orders else value for key,value in inventory.items()}

def calculate_order_statistics(customer_orders, products):
    ttl_order = len(customer_orders)
    percentage = round((ttl_order / len(products)) * 100)
    print()
    return ttl_order, percentage

def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print(f"Total Products Ordered: {order_statistics[0]}")
    print(f"Percentage of Unique Products Ordered: {order_statistics[1]}%") 
    print()
    
def print_updated_inventory(inventory):
    print("Updated Inventory:")
    inventory = dict(filter(lambda item: item[1]!=0, inventory.items()))
    for key, value in inventory.items():
        print(f"{key}: {value}")
    print()
        
def calculate_total_price(customer_orders):
    total_price = 0
    for order in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = int(input(f"Enter the price of {order}: "))
                if price < 0:
                    raise ValueError("Invalid price! Please enter a non-negative value.")
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
        total_price += price
    return print(f"Total Price: {total_price}")
        
    
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
calculate_total_price(customer_orders)
````
